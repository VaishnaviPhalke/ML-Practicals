C) Linear Regression

```python
X1 = df1.drop("Outcome", axis=1)
y1 = df1["Outcome"]

X_train1, X_test1, y_train1, y_test1 = train_test_split(
    X1, y1, test_size=0.2, random_state=42)

lin_model1 = LinearRegression()
lin_model1.fit(X_train1, y_train1)

y_pred_lin1 = lin_model1.predict(X_test1)

print("R2 Score:", r2_score(y_test1, y_pred_lin1))
print("RMSE:", np.sqrt(mean_squared_error(y_test1, y_pred_lin1)))
```

D) Logistic Regression (Correct Model)

```python
log_model1 = LogisticRegression(max_iter=1000)
log_model1.fit(X_train1, y_train1)

y_pred_log1 = log_model1.predict(X_test1)

print("Accuracy:", accuracy_score(y_test1, y_pred_log1))
print("Confusion Matrix:\n", confusion_matrix(y_test1, y_pred_log1))
print("\nClassification Report:\n", classification_report(y_test1, y_pred_log1))
```

E) Multiple Regression (Using 3 Strong Features)

```python
X_multi1 = df1[['Glucose', 'BMI', 'Age']]
y_multi1 = df1['Outcome']

multi_model1 = LinearRegression()
multi_model1.fit(X_multi1, y_multi1)

print("Coefficients:", multi_model1.coef_)
print("Intercept:", multi_model1.intercept_)
```

A) Preprocessing

```python
df2['class'] = df2['class'].map({'Positive':1, 'Negative':0})
```

Encode categorical columns:

```python
le = LabelEncoder()
for col in df2.columns:
    df2[col] = le.fit_transform(df2[col])
```

D) Logistic Regression

```python
X2 = df2.drop("class", axis=1)
y2 = df2["class"]

X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X2, y2, test_size=0.2, random_state=42)

log_model2 = LogisticRegression(max_iter=1000)
log_model2.fit(X_train2, y_train2)

y_pred_log2 = log_model2.predict(X_test2)

print("Accuracy:", accuracy_score(y_test2, y_pred_log2))
print("Confusion Matrix:\n", confusion_matrix(y_test2, y_pred_log2))
print("\nClassification Report:\n", classification_report(y_test2, y_pred_log2))
```
